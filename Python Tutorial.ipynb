```python
import math
def cosine_similarity(v1,v2):
    "compute cosine similarity of v1 to v2: (v1 dot v2)/{||v1||*||v2||)"
    sumxx, sumxy, sumyy = 0, 0, 0
    for i in range(len(v1)):
        x = v1[i]; y = v2[i]
        sumxx += x*x
        sumyy += y*y
        sumxy += x*y
    return sumxy/math.sqrt(sumxx*sumyy)

v1,v2 = [3, 45, 7, 2], [2, 54, 13, 15]
print(v1, v2, cosine_similarity(v1,v2))
```

```python
import math
def Cos(q, d):
    qq_sum, qd_sum, dd_sum = 0, 0, 0
    for i in range(len(q)):
        a = q[i]
        b = d[i]
        qq_sum += a*a
        qd_sum += a*b
        dd_sum += b*b
    return qd_sum / math.sqrt(qq_sum * dd_sum)

q = []
d = []
n = int(input())
for i in range(n):
    q.append(int(input()))
    
for i in range(n):
    d.append(int(input()))
    
print(Cos(q,d))
```
